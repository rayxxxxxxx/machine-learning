```python
import math
import random

import numpy as np

from matplotlib import pyplot as plt
from matplotlib.pyplot import cm
```

```python
class Model1:
    def __init__(self, p: float, a: float) -> None:
        self.p = p
        self.a = a
        self.s = 0.0

    def update(self, x):
        self.s = self.p*self.s+math.exp(-self.s)*x

    def activation(self):
        return self.s
```

```python
class Model2:
    def __init__(self, p: float, a: float) -> None:
        self.p = p
        self.a = a
        self.t = 0.0
        self.s = 0.0

    def update(self, x):
        u = 1/(1+math.exp(-x))
        self.s = self.p*self.s+u
        self.t = self.a*self.t+u

    def activation(self):
        return self.s*math.exp(-self.t)
```

```python
p = 0.9
a = 0.5
```

```python
# model = Model1(p ,a)
model = Model2(p ,a)

fig = plt.figure(figsize=(10, 5), dpi=100)
ax = fig.add_subplot()

y = []
nsteps = 500
for i in range(nsteps):
    y.append(model.activation())

    # x = 1
    # x = 1/(i+1)
    x = math.sin(0.1*i)
    # x = 1 if i!=0 and i == nsteps//4 else 0
    # x = 1 if i%(nsteps//4)==0 else 0

    model.update(x)
    
ax.plot(range(nsteps), y, c='black')

plt.tight_layout()
plt.show()
```
